# 02 — Reproduce Su et al. 2019 baselines (N=274)

Train simple multimodal baselines:
- **Regression:** predict each constitution converted score from biosignal features
- **Classification:** predict derived primary constitution type

Paper used linear regression + Pearson correlation; we use Random Forest for a stronger sklearn baseline on this small cohort.

```python
from tcm_constitution.adapters.mendeley_274 import load_mendeley_274
from tcm_constitution.models.mendeley_baseline import results_to_frame, train_mendeley_baselines

cohort = load_mendeley_274()
results = train_mendeley_baselines(cohort)
results_to_frame(results)
```

```python
import pandas as pd
from tcm_constitution.ccmq.scoring import CONSTITUTION_LABELS

mae = pd.Series(results.regression_mae, name="MAE").rename(index=CONSTITUTION_LABELS)
r2 = pd.Series(results.regression_r2, name="R2").rename(index=CONSTITUTION_LABELS)
pd.concat([mae, r2], axis=1).round(3)
```

```python
print(f"Accuracy: {results.classification_accuracy:.3f}")
print(f"Macro-F1: {results.classification_macro_f1:.3f}")
print(f"Split: {results.n_train} train / {results.n_test} test")
```

## Interpretation

- MAE on 0–100 scores shows how well tongue+voice+pulse indices recover constitution subscales
- Macro-F1 accounts for imbalanced primary types (see notebook 01)
